```python
import re
from google.colab import drive

# Google 드라이브를 마운트합니다.
drive.mount('/content/drive')

# 'Avengers.txt' 파일을 열고 내용을 읽은 다음, 모든 문자를 소문자로 변환합니다.
with open('/content/drive/MyDrive/Colab Notebooks/Avengers.txt', 'r') as file:
    file_content = file.read().lower()

# 모든 기호를 제거합니다.
cleaned_content = re.sub(r'[^\w\s]', '', file_content)

# 결과를 출력합니다.
print(cleaned_content)
```

```python
import re
from collections import Counter

# 원본 텍스트를 소문자로 변환하고 기호를 제거합니다.
text = "the tesseract has awakened it is on a neutral world a human world they wield its power but our ally knows its working so that they never will learn hes ready to lead and our force our chitauri will follow the world will be his and the universe yours and the humans what can they do but burn".lower()
cleaned_text = re.sub(r'[^\w\s]', '', text)

# 2-gram을 구합니다.
words = cleaned_text.split()
bigrams = zip(words, words[1:])
bigram_counts = Counter(bigrams)

# 가장 많이 나오는 2-gram 쌍을 찾습니다.
most_common_bigram = bigram_counts.most_common(1)

print(most_common_bigram)
```
